# **Create an Excel Sheet containing marks of 10 students for 5 subjects ot of 100 using Pandas read the excel file and do the result analysis.**

```python
import pandas as pd
import openpyxl
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel (r'/content/Marks.xlsx')
data = pd.DataFrame(data)
```

```python
data1 = data.copy()
```

# a) Total Marks of Each Student

```python
Total = data1.sum(axis = 1)
Total = pd.DataFrame(Total)
data1.insert(6,'TOTAL',Total)
data1
```

# b) Average of Individual student

```python
data1 = data1.drop(['TOTAL'], axis = 1)
avg = data1.mean(axis = 1)
avg = pd.DataFrame(avg)
data1.insert(6, 'TOTAL', Total)
data1.insert(7, 'AVERAGE', avg)
data1
```

# c) Number of Students passed and Failed in each Subject

```python
pp = [1,2,3,4,5]
sum = 0
for col in range(1,6):
  for row in range(0,10):
    if data[data.columns[col]][row] > 40:
      sum = sum + 1
  pp[col-1] = sum
  sum = 0
```

```python
fp = [10,10,10,10,10]
array1 = np.array(fp)
array2 = np.array(pp)
subtracted_array = np.subtract(array1, array2)
subtracted = list(subtracted_array)
```

```python
p_f = {'SUBJECT': data.columns[1:6] ,'NO. OF STUDENTS PASSED' : pp , 'NO. OF STUDENTS FAILED' : subtracted}
p_f = pd.DataFrame(p_f)
p_f
```

# d) Find the Pass Percentage and Fail Percentage of each subject (plotting Pie Chart)

```python
pp = [1,2,3,4,5]
sum = 0
for s in range(1,6):
  for i in range(0,10):
    if data[data.columns[s]][i] > 40:
      sum = sum + 1
  pp[s-1] = (sum/10)*100
  sum = 0

fp = [100,100,100,100,100]
array3 = np.array(fp)
array4 = np.array(pp)
subtracted_array = np.subtract(array3, array4)
subtracted = list(subtracted_array)

data1 = data.drop(['NAME'], axis = 1)
pass_percentage = {'SUBJECT': data1.columns ,'PASS PERCENTAGE' : pp, 'FAIL PERCENTAGE' : subtracted}
pass_percentage = pd.DataFrame(pass_percentage)
```

```python
p = np.array([pp[0], pp[1], pp[2], pp[3], pp[4]])
label1 = ['DSP', 'ANN', 'AAS', 'LSE','CODING']
plt.pie(p, labels = label1, startangle = 90)
plt.title('PASS PERCENTAGE')
# plt.legend()
plt.show()

print('\n')

f = np.array([subtracted[0], subtracted[1], subtracted[2], subtracted[3], subtracted[4]])
label2 = ['DSP', 'ANN', 'AAS', 'LSE','CODING']
plt.pie(f, labels = label2, startangle = 90)
plt.title('FAIL PERCENTAGE')
# plt.legend()
plt.show()
```

# e) Find the number of students who have secured the marks in range 80-100 in a particular subject.

```python
t = [0,0,0,0,0]
for c in range(0,5):
  sum = 0
  for r in range(0,10):
    if data1[data1.columns[c]][r] > 80:
      sum = sum + 1
  t[c]=sum

t
```

```python
g80 = {'STUDENT NAME': data1.columns ,'No. of Students  scored > 80' : t}
g80 = pd.DataFrame(g80)
g80
```
